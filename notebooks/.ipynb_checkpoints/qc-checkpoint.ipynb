```python
def get_raw_readcounts(metadata):
    """
    Function takes in the metadata and outputs the sum of read counts for the given experiment/s
    """
    readcounts = metadata.groupby("exp")["read_count"].sum().to_dict()
    run_types = metadata.groupby("exp")["run_type"].first().to_dict()
    for exp in readcounts:
        if run_types[exp] == "paired-ended":
            readcounts[exp] = readcounts[exp] // 2 # divide by 2 for paired-end to count read pairs
    
    return readcounts
```

```python
def get_trimmed_readcounts(experiment, pe_log_dir, se_log_dir):
    """
    Function takes in the experiment list, and ouput directories for the trimming rule and outputs
    the sum of read counts for the given experiment/s.
    """
    pe_log_files = [f for f in os.listdir(pe_log_dir) if f.startswith(experiment) and f.endswith('.log')]
    se_log_files = [f for f in os.listdir(se_log_dir) if f.startswith(experiment) and f.endswith('.log')]
    total_surviving = 0

    for log_file in pe_log_files + se_log_files:
        log_dir = pe_log_dir if log_file in pe_log_files else se_log_dir
        with open(os.path.join(log_dir, log_file), 'r') as f:
            content = f.read()
            match = re.search(r'Surviving: (\d+) \(([\d.]+)%\)', content)
            if match:
                surviving = int(match.group(1))
                total_surviving += surviving  # Trimmomatic log file counts surviving read pairs, not individial reads (don't need to divide by 2)

    return total_surviving
```

```python
def get_aligned_readcounts(experiment, pe_bam_dir, se_bam_dir, genomes, metadata):
    """
    Function takes in the experiment list, and ouput directories for the bowtie2 alignment rule and outputs
    the sum of read counts for the given experiment/s.
    """
    aligned_counts = {}
    exp_metadata = metadata[metadata['exp'] == experiment]
    bioreps = exp_metadata['biorep'].unique()
    is_paired_end = exp_metadata['run_type'].iloc[0] == 'paired-ended'
    
    bam_dir = pe_bam_dir if is_paired_end else se_bam_dir
    
    for genome in genomes:
        aligned_count = 0
        for biorep in bioreps:
            bam_file = os.path.join(bam_dir, f"{genome}-{experiment}-{biorep}.bam")
            
            if os.path.exists(bam_file):
                flagstat_output = pysam.flagstat("-@", "24", bam_file)
                match = re.search(r'(\d+) \+ \d+ mapped', flagstat_output)
                if match:
                    aligned_count += int(match.group(1))
        
        if is_paired_end:
            aligned_count = aligned_count // 2  # divide by 2 for paired-end to count fragments
        
        aligned_counts[genome] = aligned_count if aligned_count > 0 else None
    
    return aligned_counts
```

```python
def get_filtered_readcounts(experiment, filtered_bam_dir, genomes, metadata):
    """
    Function takes in the experiment list, and ouput directory for the samtools filtering rule, 
    genome, and metadata and outputs the sum of read counts for the given each experiment and genome.
    """
    filtered_counts = {}
    exp_metadata = metadata[metadata['exp'] == experiment]
    is_paired_end = exp_metadata['run_type'].iloc[0] == 'paired-ended'
    
    for genome in genomes:
        bam_file = os.path.join(filtered_bam_dir, f"{genome}-{experiment}.bam")
        
        if os.path.exists(bam_file):
            flagstat_output = pysam.flagstat("-@", "24", bam_file)
            match = re.search(r'(\d+) \+ \d+ mapped', flagstat_output)
            if match:
                filtered_count = int(match.group(1))
                filtered_counts[genome] = filtered_count
        else:
            filtered_counts[genome] = None
    
    return filtered_counts
```

```python
def get_kmer_filtered_readcounts(experiment, kmer_filtered_bam_dir, genomes, metadata):
    """
    Function takes in the experiment list, and ouput directory for the unique kmer filtering rule, 
    genome, and metadata and outputs the sum of read counts for the given each experiment and genome.
    """
    kmer_filtered_counts = {}
    exp_metadata = metadata[metadata['exp'] == experiment]
    is_paired_end = exp_metadata['run_type'].iloc[0] == 'paired-ended'
    
    for genome in genomes:
        bam_file = os.path.join(kmer_filtered_bam_dir, f"{genome}-{experiment}.bam")
        
        if os.path.exists(bam_file):
            flagstat_output = pysam.flagstat("-@", "24", bam_file)
            match = re.search(r'(\d+) \+ \d+ mapped', flagstat_output)
            if match:
                kmer_filtered_count = int(match.group(1))
                kmer_filtered_counts[genome] = kmer_filtered_count
        else:
            kmer_filtered_counts[genome] = None
    
    return kmer_filtered_counts
```

```python
# Directory paths
pe_log_directory = '/zata/zippy/ramirezc/scratch/logs/trim_pe/'
se_log_directory = '/zata/zippy/ramirezc/scratch/logs/trim_se/'
pe_bam_directory = '/zata/zippy/ramirezc/scratch/aln-raw-pe/'
se_bam_directory = '/zata/zippy/ramirezc/scratch/aln-raw-se/'
filtered_bam_directory = '/zata/zippy/ramirezc/scratch/aln-filtered/'
kmer_filtered_bam_directory = '/zata/zippy/ramirezc/scratch/aln-final/'

genomes = ['GRCh38', 'T2T']
```

```python
# Process readcounts (multi-threaded pysam.flagstat ftw!)
raw_readcounts = get_raw_readcounts(metadata[metadata['exp'].isin(experiments)])
df_raw = pd.DataFrame(raw_readcounts.items(), columns=['exp', 'raw_read_count'])

trimmed_readcounts = {}
aligned_readcounts = {}
filtered_readcounts = {}
kmer_filtered_readcounts = {}

for exp in experiments:
    trimmed_readcounts[exp] = get_trimmed_readcounts(exp, pe_log_directory, se_log_directory)
    aligned_readcounts[exp] = get_aligned_readcounts(exp, pe_bam_directory, se_bam_directory, genomes, metadata)
    filtered_readcounts[exp] = get_filtered_readcounts(exp, filtered_bam_directory, genomes, metadata)
    kmer_filtered_readcounts[exp] = get_kmer_filtered_readcounts(exp, kmer_filtered_bam_directory, genomes, metadata)

df_trimmed = pd.DataFrame(trimmed_readcounts.items(), columns=['exp', 'trimmed_read_count'])
df_aligned = pd.DataFrame(aligned_readcounts).T.reset_index()
df_aligned.columns = ['exp'] + genomes
df_filtered = pd.DataFrame(filtered_readcounts).T.reset_index()
df_filtered.columns = ['exp'] + genomes
df_kmer_filtered = pd.DataFrame(kmer_filtered_readcounts).T.reset_index()
df_kmer_filtered.columns = ['exp'] + genomes
```

```python
# Concatenate all dataframes
df_final = df_raw.merge(df_trimmed, on='exp', how='outer')

for genome in genomes:
    df_final = df_final.merge(
        df_aligned[['exp', genome]].rename(columns={genome: f'aligned_{genome}'}),
        on='exp', how='outer'
    )
    df_final = df_final.merge(
        df_filtered[['exp', genome]].rename(columns={genome: f'filtered_{genome}'}),
        on='exp', how='outer'
    )
    df_final = df_final.merge(
        df_kmer_filtered[['exp', genome]].rename(columns={genome: f'kmer_filtered_{genome}'}),
        on='exp', how='outer'
    )
```
